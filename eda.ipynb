```python
import numpy as np
import pandas as pd
```

```python
customers_df = pd.read_csv('datasets/olist_customers_dataset.csv')
customers_df
```

```python
orders_df = pd.read_csv('datasets/olist_orders_dataset.csv')
orders_df
```

```python
order_review_df = pd.read_csv('datasets/olist_order_reviews_dataset.csv')
order_review_df
```

```python
# join the order and order review dataframes
merged_df = pd.merge(orders_df, order_review_df, on='order_id')

# join the merged dataframe with the customers dataframe
merged_df = pd.merge(merged_df, customers_df, on='customer_id')

merged_df
```
