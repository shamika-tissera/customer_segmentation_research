```python
import numpy as np
import pandas as pd

customer_reviews_df = pd.read_csv('datasets/olist_order_reviews_dataset.csv')
customer_reviews_df
```

```python
output_df = customer_reviews_df.copy()
for index, row in customer_reviews_df.iterrows():
    if customer_reviews_df.loc[index, 'review_comment_message'] is np.nan:
        output_df.loc[index, 'review_comment_message_en'] = np.nan
    else:
        output_df.loc[index, 'review_comment_message_en'] = translator.translate(row['review_comment_message'], dest='en').text
        
    if customer_reviews_df.loc[index, 'review_comment_title'] is np.nan:
        output_df.loc[index, 'review_comment_title_en'] = np.nan    
    else:
        output_df.loc[index, 'review_comment_title_en'] = translator.translate(row['review_comment_title'], dest='en').text
    
    print(output_df.loc[index, 'review_comment_message_en'])
    output_df.to_csv('datasets/olist_order_reviews_translated.csv', index=False)
```
